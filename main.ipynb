```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("Canada_per_capita_income.csv")  # Replace with your actual filename
```

```python
# Optional: Check the data
print("Data Preview:")
print(df.head())
```

```python
# Correct the typo
df = pd.DataFrame()
df
```

```python
# Ensure the DataFrame is loaded with data before describing it
if not df.empty:
	df.info()
	print(df.describe())
else:
	print("The DataFrame is empty. Please load data into 'df' before describing it.")
```

```python
# Extract features (X) and labels (y)
X = df[['year']]  # Must be a 2D array
y = df['income']  # Replace column name if different
```

```python
# Step 3: Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict income for 2020
predicted_2020 = model.predict([[2020]])
print(f"\nPredicted per capita income in 2020: ${predicted_2020[0]:,.2f}")
```

```python
# Plot the data and prediction
plt.scatter(X, y, color='blue', label='Training data')
plt.plot(X, model.predict(X), color='green', label='Regression line')
plt.scatter([2020], predicted_2020, color='red', label='Prediction (2020)', marker='x', s=100)

plt.title('Canada Per Capita Income Prediction')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.legend()
plt.grid(True)
plt.show()
```
